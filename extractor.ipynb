```python
def translate(seq):
      
    table = {
        'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
        'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
        'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
        'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',                
        'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
        'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
        'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
        'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
        'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
        'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
        'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
        'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
        'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
        'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
        'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_',
        'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W',
    }
    protein = ""
    for i in range(0, len(seq)//3*3, 3):
        codon = seq[i:i + 3]
        protein+= table[codon]
    return protein

import os
dir = '/home/ars1suhod/genbank_bacterial_curlins' # указываем нашу директорию

l = []
for path, subdirs, files in os.walk('/home/ars1suhod/genbank_bacterial_curlins'):
    for filename in files:
        if filename.endswith('.txt'):
            l.append(filename)
#l = ['1','2','3'] #имена файлов
with open('extractor.txt', 'w') as f:
    for i in l:
        f.write('>' + i.split(' ')[0] + '\n')
        k = 0
        l = 0
        s = ''
        with open('/home/ars1suhod/genbank_bacterial_curlins/' + i, 'r') as fd:
            for line in fd:
                k+=1
                if line.find('protein=curli') != -1:
                    print(k)
                    break
        with open('/home/ars1suhod/genbank_bacterial_curlins/' + i, 'r') as fd:
            for line in fd:
                l+=1
                if l >= (k - 170) and l <= (k + 170):
                    #print(l)
                    if line.find('protein') == -1:
                        s = s + line
            #print(s)
            s = s.replace('\n', '')
            s = translate(s)
            print(s)
            f.write(s)
        f.write('\n')
                
            
```
