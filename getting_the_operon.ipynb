```python
import numpy as np

import pandas

data = pandas.read_csv('new_curli_bact_contigs_summary.tsv', delimiter='\t')
print(data)
```

```python
candidates = data[['contig_id', 'cds_id', 'start', 'end']]
print(candidates)
```

```python
with open ('operons.txt', 'w') as f:
    f.write('operon 1')
    f.write('\n')
    f.write(candidates['cds_id'][0])
    f.write('\n')
    j = 1
    for i in range (1, len(candidates)):
        if (candidates['start'][i] - candidates['end'][i-1] <= 20) and (candidates['start'][i] - candidates['end'][i-1] >= 0) :
            f.write(candidates['cds_id'][i])
            f.write('\n')
        else:
            j+=1
            f.write('\n')
            f.write('operon' + ' ' + str(j))
            f.write('\n')
            f.write(candidates['cds_id'][i])
            f.write('\n')
```

```python
s = []
with open ('results.domtblout', 'r') as g:
    for line in g:
        k = line.split(' ')
        for element in k:
            if 'cds' in element:
                s.append(element)
print(s)
```

```python
l = []
with open('operons.txt', 'r') as f:
    for line in f:
        counter = 0
        if 'operon' in line:
            a = line.split('\n')[0]
        if line.split('\n')[0] in s:
            l.append(a)
curli_operon_list = list(set(l))
print(l)
```

```python
with open('operons.txt', 'r') as f:
    with open('curli_operons.txt', 'w') as g:
        counter = 0
        for line in f:
            if 'operon' in line and line.split('\n')[0] in curli_operon_list:
                counter = 1
                g.write('\n')
            if 'operon' in line and line.split('\n')[0] not in curli_operon_list:
                counter = 0
            if counter == 1:
                g.write(line)
```

```python
curli_list = []
with open('operons.txt', 'r') as f:
    counter = 0
    for line in f:
        if counter == 1:
            curli_list.append('>' + line.split('|')[0])
        if 'operon' in line and line.split('\n')[0] in curli_operon_list:
            counter = 1
        if 'operon' in line and line.split('\n')[0] not in curli_operon_list:
            counter = 0
print(curli_list)
```

```python
with open ('sequences_bacteria_curli_aa.faa', 'r') as f:
    with open ('curli_operons.faa', 'w') as g:
        counter = 0
        for line in f:
            if counter == 1:
                g.write(line)
            if any(element in line.split('\n') for element in curli_list):
                g.write(line)
                counter = 1
            else:
                counter = 0
            
```
