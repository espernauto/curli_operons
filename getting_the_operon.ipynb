```python
def extract(specie):
    data = pandas.read_csv(specie + '.tsv', delimiter='\t')
    print(data)
    candidates = data[['contig_id', 'cds_id', 'start', 'end']]
    s = []
    with open (specie + '.domtblout', 'r') as g:
        for line in g:
            k = line.split(' ')
            for element in k:
                if 'cds' in element:
                    s.append(element)
    print(s)
    l = []
    with open('operons.txt', 'r') as f:
        for line in f:
            counter = 0
            if 'operon' in line:
                a = line.split('\n')[0]
            if line.split('\n')[0] in s:
                l.append(a)
    curli_operon_list = list(set(l))
    curli_list = []
    with open('operons.txt', 'r') as f:
        counter = 0
        for line in f:
            if counter == 1:
                curli_list.append('>' + line.split('|')[0])
            if 'operon' in line and line.split('\n')[0] in curli_operon_list:
                counter = 1
            if 'operon' in line and line.split('\n')[0] not in curli_operon_list:
                counter = 0
    print(curli_list)
    with open (specie + '.faa', 'r') as f:
        with open ('curli_operons.faa', 'w') as g:
            counter = 0
            for line in f:
                if counter == 1:
                    g.write(line)
                if any(element in line for element in curli_list):
                    g.write(line)
                    counter = 1
                else:
                    counter = 0
```
